```python
cur_x=3
lr=0.0001
precision=0.000001
max_iter=10000
iters=0
p_step_size=1
df=lambda x:2*(x+5)

while p_step_size>precision and iters < max_iter:
    prev_x=cur_x
    cur_x=cur_x-lr*df(prev_x)
    p_step_size=abs(cur_x-prev_x)
    iters=iters+1
    print("iterations",iters,"\n X value is",cur_x)

print("the local minima is ",cur_x)
```
